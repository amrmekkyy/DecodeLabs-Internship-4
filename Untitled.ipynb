```python
# import libraries
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing import image_dataset_from_directory
import matplotlib.pyplot as plt
```

```python
# Dataset path
dataset_path = "fruit_and_vegetables"
```

```python
# Load training dataset
train_ds = image_dataset_from_directory(
    dataset_path,
    image_size=(224, 224),
    batch_size=32
)
```

```python
# Display classes names
class_names = train_ds.class_names
print("Classes:", class_names)
```

```python
# Build the classification model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(8, activation='softmax')
])
```

```python
# Preparing the model for training
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train model
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30
)
```

```python
# estimate train and validation model accuracy
plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train'])
plt.show()
```

```python
import os
import random

# List to store all image paths
all_images = []

# Loop over each class folder
for class_name in class_names:
    
    # Path of current class folder
    class_folder = os.path.join(dataset_path, class_name)
    
    # Loop over images inside folder
    for img_name in os.listdir(class_folder):
        
        # Full image path
        img_path = os.path.join(class_folder, img_name)
        
        # Add image path to list
        all_images.append(img_path)

# Randomly select 5 images from dataset
random_images = random.sample(all_images, 8)

def predict_and_show(img_path):
    
    # Load image for model
    img = tf.keras.utils.load_img(
        img_path,
        target_size=(224, 224)
    )
    
    # Convert image to array
    img_array = tf.keras.utils.img_to_array(img)
    
    # Add batch dimension
    img_array = tf.expand_dims(img_array, 0)
    
    # Make prediction
    prediction = model.predict(img_array)
    
    # Get predicted class name
    predicted_class = class_names[tf.argmax(prediction[0])]
    
    # ===== Show Image =====
    plt.imshow(img)
    plt.axis('off')
    
    # Show title with prediction
    plt.title(f"Prediction: {predicted_class}")
    plt.show()
    
    # Print path (optional debug)
    print("Image Path:", img_path)
    print("Predicted Class:", predicted_class)
    print("-" * 40)

# Loop over random images
for img_path in random_images:
    predict_and_show(img_path)
```
